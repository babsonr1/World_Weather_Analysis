```python
import random
import numpy as np
import timeit
import pandas as pd
import matplotlib.pyplot as plt
from citipy import citipy
import requests
from config import weather_api_key
from datetime import datetime
import time
from scipy.stats import linregress
```

```python
lats = np.random.uniform(low=-90.000, high=90.000, size = 1500)
longs = np.random.uniform(low=-180.000, high=180.000, size = 1500)
latlongs = zip(lats, longs)
coordinates = list(latlongs)
```

```python
for coordinate in coordinates:
    # print(citipy.nearest_city(coordinate[0], coordinate[1]).city_name,
          #citipy.nearest_city(coordinate[0], coordinate[1]).country_code)
    pass
```

```python
url = "http://api.openweathermap.org/data/2.5/weather?units=Imperial&appid=" + weather_api_key
```

```python
cities = []
for coordinate in coordinates:
    city = citipy.nearest_city(coordinate[0], coordinate[1]).city_name

    if city not in cities:
        cities.append(city)
len(cities)
```

```python
cityData = []
#cityWeather = requests.get(cityurl)
print("  Beginning Data Retrieval   ")
print("-----------------------------")

recordCount = 1
setCount = 1
```

```python
for i, city in enumerate(cities):
    if (i % 50 == 0 and i >= 50):
        setCount += 1
        recordCount = 1
    cityurl = url + "&q=" + city.replace(" ","+")
    
    print(f"Processing Record {recordCount} of Set {setCount} | {city}")
    recordCount += 1
    
    try:
        # Parse the JSON and retrieve data.
        cityWeather = requests.get(cityurl).json()
        
        # Parse out the needed data.
        cityLat = cityWeather["coord"]["lat"]
        cityLong = cityWeather["coord"]["lon"]
        cityMaxTemp = cityWeather["main"]["temp_max"]
        cityHumidity = cityWeather["main"]["humidity"]
        cityClouds = cityWeather["clouds"]["all"]
        cityWind = cityWeather["wind"]["speed"]
        cityCountry = cityWeather["sys"]["country"]
        cityDate = datetime.utcfromtimestamp(cityWeather["dt"]).strftime('%Y-%m-%d %H:%M:%S')
        cityData.append({"City": city.title(),
                          "Lat": cityLat,
                          "Lng": cityLong,
                          "Max Temp": cityMaxTemp,
                          "Humidity": cityHumidity,
                          "Cloudiness": cityClouds,
                          "Wind Speed": cityWind,
                          "Country": cityCountry,
                          "Date": cityDate})

    # If an error is experienced, skip the city.
    except:
        print("City not found. Skipping...")
        pass

# Indicate that Data Loading is complete.
print("-----------------------------")
print("Data Retrieval Complete      ")
print("-----------------------------")
```

```python
cityDatadf = pd.DataFrame(cityData)
cityDatadf = cityDatadf[["City","Country","Date","Lat","Lng", "Max Temp", "Humidity", "Cloudiness", "Wind Speed"]]
cityDatadf.head()
```

```python
def plotLinearRegression(x_values, y_values, title, y_label, text_coordinates):
    (slope, intercept, r_value, p_value, std_err) = linregress(x_values, y_values)
    
    regressValues = x_values*slope + intercept
    lineEquation = "y = " + str(round(slope,2)) + "x + " + str(round(intercept,2))
    
    plt.scatter(x_values, y_values)
    plt.plot(x_values, regressValues, "r")
    
    plt.annotate(lineEquation, text_coordinates, fontsize=15, color = "red")
    plt.title(title)
    plt.xlabel("Latitude")
    plt.ylabel(y_label)
    
    plt.show()
```

```python
northernHemidf = cityDatadf.loc[(cityDatadf["Lat"] >= 0)]
southernHemidf = cityDatadf.loc[(cityDatadf["Lat"] < 0)]
```

```python
northernXValues = northernHemidf["Lat"]
northernMaxTempValues = northernHemidf["Max Temp"]

plotLinearRegression(northernXValues, northernMaxTempValues, "Linear Regression on the Northern Hemisphere \n for Maximum Temperature", "Max Temp", (5,20))
```
